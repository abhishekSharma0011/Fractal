```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-50, 50, 1000)
y = x**2

plt.plot(x,y)
plt.title("y=x^2")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.grid(True)
plt.axhline(100, color="green", linewidth=1)
plt.axvline(20, color="red", linewidth=1)
plt.show()
```

```python

y = np.sin(x)

plt.plot(x,y)
plt.title("Sine Wave: y = sin(x)")
plt.show()
```

```python
y = np.exp(x)

plt.plot(x,y)
plt.title("Exponent function: y = exp(x)")
plt.show()
```
