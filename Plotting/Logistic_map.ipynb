```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# parameters
r = 3.5
x = 0.6
iterations = 100
xs = []
```

```python
for _ in range(iterations):
    x = r*x*(1-x)
    xs.append(x)
    
```

```python
plt.plot(xs, marker='o')
plt.title(f'Logistic map (r={r}), (x={x})')
plt.xlabel('Iteration')
plt.ylabel('X')
plt.show()
```
