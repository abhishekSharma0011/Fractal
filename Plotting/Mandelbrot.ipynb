```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Image size
width, height = 800, 800

# create complex plane
real = np.linspace(-2, 1, width)
imag = np.linspace(-1.5, 1.5, height)
mandelbrot = np.zeros((height, width))
```

```python
# Max iterations
max_iter = 80

for i in range(height):
    for j in range(width):
        c = complex(real[j], imag[i])
        z = 0
        count = 0

        while abs(z) <= 2 and count < max_iter:
            z = z*z + c
            count += 1

        mandelbrot[i, j] = count
```

```python
# plot
plt.imshow(mandelbrot, cmap='viridis', extent=(-2, 1, -1.5, 1.5))
plt.title("Mandelbrot Set")
plt.colorbar()
plt.show()
```
